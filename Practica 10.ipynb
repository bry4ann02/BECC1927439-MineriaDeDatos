```python
import wikipedia 
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
```

Se importa un texto sacado de wikipedia
Se puede extraer un resumen mediante el metodo .summary()

```python
info = wikipedia.summary("Bayern Munchen History")
info
```

```python
WCloud = WordCloud().generate(info)
```

```python
plt.imshow(WCloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
image=WCloud.to_image()
image.save('BMunich.png')
image.show()
```
