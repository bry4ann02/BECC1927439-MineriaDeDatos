```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#Se lee el dataset que usaremos
ruta = "C:/Users/becar/OneDrive/Escritorio/Mineria de Datos/BECC1927439-MineriaDeDatos/mls_salaries_masterlist.csv"
data = pd.read_csv(ruta)

df = pd.DataFrame(data)
df
```

Se obtiene la media de salarios en las dos conferencias Western y Eastern 

```python
western = df[df['conference'] == 'Western']['base_salary'].mean()

eastern = df[df['conference'] == 'Eastern']['base_salary'].mean()

western, eastern
```

Hacemos una grafica en donde se va acomparar las media de salarios de las dos conferencias 

```python
plt.figure(figsize=(8, 6))
plt.bar(['Base Salary Western', 'Base Salary Eastern'], [western, eastern], color=['blue', 'black'])
plt.xlabel('Promedio de Salario')
plt.ylabel('Conferencia')
plt.title('Diferencia de Salarios entre Conferencias')
plt.show()
```

Se obtienen las medias de las regalias en cada conferencia 

```python
western = df[df['conference'] == 'Western']['guaranteed_compensation'].mean()

eastern = df[df['conference'] == 'Eastern']['guaranteed_compensation'].mean()

western, eastern
```

Se hace las grafica de la compracion de medias de regalias

```python
plt.figure(figsize=(8, 6))
plt.bar(['Base Salary Western', 'Base Salary Eastern'], [western, eastern], color=['green', 'gray'])
plt.xlabel('promedio de Regalias')
plt.ylabel('Conferencia')
plt.title('Diferencia de regalias entre Conferencias')
plt.show()
```
